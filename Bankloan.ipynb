```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import StackingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('loans_updated.csv')
```

```python
# One Hot Encoding
df = pd.get_dummies(
    df,
    columns=[
        'Gender',
        'Education',
        'Marital_Status',
        'Employment_Status',
        'State',
        'Loan_Types',
        'Home_Ownership',
        'Mortgage',
        'Credit_History',
        'Property_Area'
    ],
    drop_first=True
)
```

```python
df.columns
```

```python
df.head()
```

```python
df.shape
```

```python
X = df.drop(['Loan_Approveled'],axis=1)
y = df['Loan_Approveled']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
df['Previous_Defaults'].unique()
```

```python
#Scaling
from sklearn.preprocessing import StandardScaler
num_cols = [
    'Age','Annual_Income','Co_Applicant_Income',
    'Savings_Balance','Existing_Loans',
    'Loan_Amount','Loan_Term',
    ]
scaler = StandardScaler()
```

```python
df.head()
```

```python
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
```

```python
y_test
```

```python
#Stacking
base_learner = [
    ('dt', DecisionTreeClassifier(random_state=42)),
    ('svm',SVC(probability=True, kernel='rbf',random_state=42)),
    ('lr',LogisticRegression(max_iter=1000))
]
```

```python
meta_learner = LogisticRegression(max_iter=1000)
```

```python
stacking_CLF = StackingClassifier(
    estimators = base_learner,
    final_estimator = meta_learner,
    cv= 5
)
```

```python
stacking_CLF.fit(X_train, y_train)
```

```python
y_pred = stacking_CLF.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
#Bagging
from sklearn.ensemble import RandomForestClassifier
```

```python
model_RBF = RandomForestClassifier(
    n_estimators = 100,
    max_depth = None,
    random_state=42
)
```

```python
model_RBF.fit(X_train, y_train)
```

```python
y_pred = model_RBF.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
#Boosting
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
```

```python
model_ADA = AdaBoostClassifier(n_estimators=100, random_state=42)
```

```python
model_ADA.fit(X_train, y_train)
```

```python
y_pred = model_ADA.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
#GradientBoostingClassifier
model_GB = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
```

```python
model_GB.fit(X_train, y_train)
```

```python
y_pred = model_GB.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
